**Bar Chart Analysis for Categorical Column**

```python
categorical_df = pd.DataFrame(df.select_dtypes(include=object)).columns
categorical_df
```

```python
for col in categorical_df:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=df[col])
    #df[col].value_counts().plot(kind='bar')
    plt.title(f'Frequency of Categories in {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.show()
```

# Standardization of Variable #

```python
scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(numeric_df), columns=numeric_df.columns)



```

**Plot before and after distributions**

```python
for col in numeric_df.columns:
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(numeric_df[col], kde=True)
    plt.title(f'Before Standardization: {col}')
    
    plt.subplot(1, 2, 2)
    sns.histplot(df_standardized[col], kde=True)
    plt.title(f'After Standardization: {col}')
    
    plt.show()
```

**Conversion of Categorical Data into Dummy Variables**

```python
df_encoded = pd.get_dummies(df, columns=categorical_df)

# Display a portion of the transformed dataset
print("Transformed Dataset:")
print(df_encoded.head())
```

**Combine standardize numerical and encoded categorical data**

```python
processed_df = pd.concat([df_standardized, df_encoded], axis=1)

processed_df.head()
```

Save the processed Dataset to a new CSV file

```python
processed_df.to_csv('processed_sales_discounts.csv', index=False)
```
